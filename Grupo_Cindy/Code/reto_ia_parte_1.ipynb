```python
import pandas as pd
from bs4 import BeautifulSoup
import requests as req
from unidecode import unidecode
```

```python
df_table = pd.read_csv('all_data.csv', sep=',', encoding= 'ISO-8859-1')
df_table.referencia = df_table.referencia.map(lambda x: unidecode(x).lower().strip())
```

```python
with open(r'soup.html', "r") as f:
    page = f.read()
S = BeautifulSoup(page, 'lxml')
```

```python
total = 0
dict_ = {
    'referencia': [],
    'descripcion': [],
    'titulo': [],
}

for td in S.find_all('td'):
    print(td.get_text())
    if 'fdsRequestSummaryInfo_tblDetail_trRowRef_' in td.get('id'):
        dict_['referencia'].append(unidecode(td.get_text().lower().strip()).replace("numero del proceso",""))
    if 'fdsRequestSummaryInfo_tblDetail_trRowDescription_' in td.get('id'):
        dict_['descripcion'].append(unidecode(td.get_text().lower()).replace("descripcion:",""))
    if 'fdsRequestSummaryInfo_tblDetail_trRowName_' in td.get('id'):
        dict_['titulo'].append(unidecode(td.get_text().lower()).replace("titulo:",""))

dict_['descripcion'].extend([''] * (len(dict_['referencia'])-len(dict_['descripcion'])))
print("Total de registros ", total)
```

```python
html_df = pd.DataFrame.from_dict(dict_)

for r, c in html_df.iterrows():
    if html_df.titulo[r] == "" and html_df.descripcion[r] == "" and html_df.referencia[r] == "":
        html_df = html_df.drop(r)
```

```python
joined_df = df_table.set_index('referencia').join(html_df.set_index('referencia'), lsuffix='table', rsuffix='html')
joined_df
joined_df.to_csv("result.csv")
```
